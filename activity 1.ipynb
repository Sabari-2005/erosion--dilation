```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img=np.zeros((600,600))
    front=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='SABARI',
               org=(50,300), fontFace=front,fontScale= 5,
               color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img=load_img()
```

```python
display_img(img)
```

```python
kernel=np.ones((5,5),dtype=np.uint8)
kernel
```

```python
ero=cv2.erode(img,kernel,iterations=5)
display_img(img)
```

```python
dil=cv2.dilate(img,kernel,iterations=3)
display_img(dil)
```

```python
white=np.random.randint(low=0,high=2,size=(600,600))
```

```python
noise=(white*255)+img
```

```python
display_img(noise)
```

```python
noise_uint8=noise.astype(np.uint8)
opening=cv2.morphologyEx(noise_uint8,cv2.MORPH_OPEN,kernel)
```

```python
display_img(opening)
```

```python
black_noise=(black*(-255))+img
```

```python
display_img(black_noise)
```
